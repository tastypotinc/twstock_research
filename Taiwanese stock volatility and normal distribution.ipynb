```python
%matplotlib inline
from twstock.twstockdb.TWStockDatabase import TWSECDB
twdb = TWSECDB("/d/StockDatabase/twsecdb.scdb")
```

```python
tsmc = twdb.query_taiwan_stock_ohlcv_time_series(2330)
hh = twdb.query_taiwan_stock_ohlcv_time_series(2317)
asus = twdb.query_taiwan_stock_ohlcv_time_series(2357)
fubon = twdb.query_taiwan_stock_ohlcv_time_series(2881)
twindex = twdb.query_taiwan_index_time_series()
```

```python
tmp = tsmc.shift(1)
tmp = tmp.ix[1:]
tsmc_d2d_volat = (tsmc.close - tmp.close)/tmp.close
tmp = hh.shift(1)
tmp = tmp.ix[1:]
hh_d2d_volat = (hh.close - tmp.close)/tmp.close
tmp = asus.shift(1)
tmp = tmp.ix[1:]
asus_d2d_volat = (asus.close - tmp.close)/tmp.close
tmp = fubon.shift(1)
tmp = tmp.ix[1:]
fubon_d2d_volat = (fubon.close - tmp.close)/tmp.close
tmp = twindex.shift(1)
tmp = tmp.ix[1:]
twindex_d2d_volat = (twindex.close - tmp.close)/tmp.close
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as stats
import datetime as dt
```

```python
n, bins, patches = plt.hist(tsmc_d2d_volat.ix[-1500:], 50, normed=1, facecolor='blue', alpha=0.75)
sigma = tsmc_d2d_volat.ix[-1500:].std() 
mu = tsmc_d2d_volat.ix[-1500:].mean() 
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r-', linewidth=2)
plt.title("TSMC")
plt.show()
stats.normaltest(tsmc_d2d_volat.ix[-1500:])
```

## Check normaltest function

```python
import numpy as np
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, 1500)
n, bins, patches = plt.hist(s, 50, normed=1, facecolor='blue', alpha=0.75)
y = mlab.normpdf(bins, mu, sigma)
l = plt.plot(bins, y, 'r-', linewidth=2)
plt.title("Normal")
plt.show()
stats.normaltest(s)
```
